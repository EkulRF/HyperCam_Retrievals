```python
import numpy as np
import radis
from radis import calc_spectrum
from radis import Spectrum
import matplotlib.pyplot as plt
```

```python
def hyperCam_check(mol):
    try:
        s = calc_spectrum(850, 1290,         # cm-1
                        molecule=mol,
                        isotope='1',
                        pressure=1.101325,   # bar
                        Tgas=300,           # K
                        mole_fraction=10**(-4),
                        path_length=500,      # cm
                        databank='hitran',  # or 'hitemp', 'geisa', 'exomol'
                        )
        #s.apply_slit(400.241, 'nm')       # simulate an experimental slit
        s.plot('absorbance')
        plt.title('Long-Wave HyperCam')
    except:
        pass

    try:
        s = calc_spectrum(1205, 1350,         # cm-1
                        molecule=mol,
                        isotope='1',
                        pressure=1.101325,   # bar
                        Tgas=300,           # K
                        mole_fraction=10**(-4),
                        path_length=500,      # cm
                        databank='hitran',  # or 'hitemp', 'geisa', 'exomol'
                        )
        #s.apply_slit(400.241, 'nm')       # simulate an experimental slit
        s.plot('absorbance')
        plt.title('Methane HyperCam')
    except:
        pass
```

```python
mol = 'SO2'

hyperCam_check(mol)
```

```python
mol = 'NO2'

hyperCam_check(mol)
```
